# <font color='red'> load data of exchange rate

```python
#  exchange rate can influance the stocks price in numerous ways.
# example: if the stock have major clients/suppliers in europe it can be effected from the dollar-euro exchange rate
# i choose to take the most imporant currencies (and the biggest economies):the dollar exchange rate for euro, yen ,yuan
# and pound
# in a real assignment i would take more currencies rates

start_day = '2010-11-01'
end_day = '2017-12-22'
dollar_dollar = []
euro_dollar = []
pound_dollar = []
yen_dollar = []
yuan_dollar = []
dates_exchange_list2 = []

x = requests.get(f'https://api.exchangeratesapi.io/history?start_at={start_day}&end_at={end_day}&base=USD')
a = x.content
a = json.loads(a.decode('utf-8'))
dates_range = pd.date_range(start=start_day, end=end_day).tolist()
for d in dates_range:
    d = d.strftime('%Y-%m-%d')
    try:
        dollar_dollar.append(a['rates'][d]['USD'])
        yen_dollar.append(a['rates'][d]['JPY'])
        pound_dollar.append(a['rates'][d]['GBP'])
        euro_dollar.append(a['rates'][d]['EUR'])
        dates_exchange_list2.append(d)
#yuan maybe a new cuurency to exchange (old one, but not exchanable)        
        yuan_dollar.append(a['rates'][d]['CNY'])       
    except:
#         print('bad date: ',d)
        continue
        
df_exchangerate_data = pd.DataFrame()
df_exchangerate_data['date']= dates_exchange_list2
df_exchangerate_data['dollar_dollar'] = dollar_dollar
df_exchangerate_data['euro_dollar'] = euro_dollar
df_exchangerate_data['yen_dollar'] = yen_dollar
df_exchangerate_data['yuan_dollar'] = yuan_dollar
df_exchangerate_data['pound_dollar'] = pound_dollar

df_exchangerate_data['date'] =pd.to_datetime(df_exchangerate_data['date'], format="%Y-%m-%d")
df = df.merge(df_exchangerate_data,left_on='Date',right_on='date',how='left')
```

# <font> task 2 :create a category trend_daily_future

```python
# path = 'C:/Users/yos/Desktop/lstm/'
# df = pd.read_csv(path+'df_spark_lstm1.csv')
```

```python
#prepare features to feature important
df['gold Average bid USD'] = df['gold Average bid USD'].str.replace(',','').astype(np.float64)
df['gold Average ask USD'] = df['gold Average ask USD'].str.replace(',','').astype(np.float64)
df['gold Average ask-bid USD'] = (df['gold Average ask USD']+df['gold Average bid USD'])/2
# make y for next day and see which feature can target the next Y
df['future_close'] = df['Close'].shift(-1)

# catch the trend of the day 
# 1: stock go up since yestarday, 0: stock go down or without a change since yestarday
# (i decided no change won't be a category for himself, (it will be a very small one, if it was)
df['trend_daily'] = (df['Close']-df['Close'].shift(1)).apply(lambda x: 0 if x<=0 else 1)
# make trend for close next day and see which feature can target the next trend of next day
df['trend_daily_future'] = (df['future_close']-df['Close']).apply(lambda x: 0 if x<=0 else 1)
# df = df.drop(columns=['Unnamed: 0'])

```

# <font> task 2 features for close price (or future close)

```python
# i removed future_close and trend_daily_future because it's data from the future (data leak). features that i will not
# have in production 
X = df.drop(columns=['Close','Date','future_close','trend_daily_future'])
y = df['Close']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

#grid search 
params = {'max_depth':np.linspace(3,10,1,dtype='int'),'min_samples_leaf': np.linspace(5,80,10,dtype='int'),'random_state':[0]}
dtr = DecisionTreeRegressor()
dtr_grid = GridSearchCV(dtr,params,n_jobs=6 ,cv=5).fit(x_train,y_train)

# feature_importances
df_feature_importanc = pd.DataFrame(dtr_grid.best_estimator_.feature_importances_,index=X.columns).rename(columns={0:'importance'})
df_feature_importanc = df_feature_importanc.sort_values(by='importance',ascending=False).head(100)
plt.figure(figsize=(15,15))
plt.title('feature importnant: low (price) and high (price) most important (make sence)',fontsize=20,)
ax = sns.barplot(y=df_feature_importanc.index, x="importance", data=df_feature_importanc).tick_params(labelsize=15)
                 

print('train rmse:',rmse(dtr_grid.best_estimator_.predict(x_train),y_train))
print('test rmse:',rmse(dtr_grid.best_estimator_.predict(x_test),y_test))
print('no overfitting based on diffrence train and test')
```

# <font color='red'> building time series

```python
########build_timeseries fot train################

def build_timeseries(mat, y_col_index):
    # y_col_index is the index of column that would act as output column
    # total number of time-series samples would be len(mat) - TIME_STEPS
    
    #dim_0: rows that suppose to be with "X dimensons of time-rows" * "features" (matrix row that enter the model)
    #dim_1: number of features (price, volum, start price...)
    #TIME_STEPS: the time window
    dim_0 = mat.shape[0] - TIME_STEPS
    dim_1 = mat.shape[1]
    
    #make a matrix that enter the model
    x = np.zeros((dim_0, TIME_STEPS, dim_1))
    y = np.zeros((dim_0,))

    for i in range(dim_0):
        x[i] = mat[i:TIME_STEPS+i]
        y[i] = mat[TIME_STEPS+i, y_col_index]
    print("length of time-series x/y",x.shape,y.shape)
    return x, y

def trim_dataset(mat, batch_size):
#    trims dataset to a size that's divisible by BATCH_SIZE    
    no_of_rows_drop = mat.shape[0]%batch_size
    if(no_of_rows_drop > 0):
        return mat[:-no_of_rows_drop]
    else:
        return mat
    
TIME_STEPS = 14
BATCH_SIZE = 20
#################################################
#train data: make time series
###################################################

x_t, y_t = build_timeseries(x_train, 0)
x_t = trim_dataset(x_t, BATCH_SIZE)
y_t = trim_dataset(y_t, BATCH_SIZE)

#test data-make time series
x_test_t, y_test_t = build_timeseries(x_test, 0)
x_test_t = trim_dataset(x_test_t, BATCH_SIZE)
y_test_t = trim_dataset(y_test_t, BATCH_SIZE)
```
